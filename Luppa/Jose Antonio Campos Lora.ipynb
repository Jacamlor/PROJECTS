# Tratamiento del Dataframe para mostrar formato óptimo

##  Extraemos las dos primeras filas, estas filas muestran el formato de dato con el que queremos trabajar por fila

```python
filas_originales = df.iloc[:2] # primeras dos filas con formato correcto 
```

## Observamor que el resto de filas para que sigan el mismo formato es la fila y su sucesiva de abajo, es decir:ç
 - La 2 con la 3 unidas forman el formato adecuado 
 - La 4 con la 5, y así sucesivamente hasta la última fila

```python
# comrpobamos que el núemro de filas sin las dos primeras sigue siendo par , para poderlas unir para tener el formato de fila óptimo
df_restante = df.iloc[2:].reset_index(drop=True)
print(len(df_restante))
if len(df_restante) % 2 != 0:
    print("Advertencia: El número de filas restantes es impar. Puede haber un problema en los datos.")
```

```python
df_restante
```

## Dividimos las filas restantes por pares e impares según su índice

```python
filas_pares = df_restante.iloc[::2].reset_index(drop=True)
filas_impares = df_restante.iloc[1::2].reset_index(drop=True)
print(filas_pares)
```

```python
print(filas_impares)
```

```python
print(len(filas_pares))
print(len(filas_impares)) # comprobamos que tengan la misma longitud
```

## Transformamos todos los datos del conjunto de filas a unir al mismo tipo de dato   (texto) para manipularlo de forma más sencilla

```python
filas_pares = filas_pares.astype(str)
filas_impares = filas_impares.astype(str)
```

## Concatenamos  filas y unimos con las extraidas

```python
df_combinado = filas_pares.fillna('') + ' ' + filas_impares.fillna('') # con esta concatenación buscamos el tener el mismo formato de fila 
```

```python
df_final = pd.concat([filas_originales, df_combinado], ignore_index=True) # Aquí tenemos el DataFrame con el mismo formato por fila
```
